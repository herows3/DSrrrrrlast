## 1. 

실제 모수 $\mu=0.5$인 베르누이 확률 변수의 시뮬레이션을 통해 평균이 0.5인지를 검정해보아라. (scipy.stats의 내장함수를 이용하되, 식으로 계산한 값과 비교하여라. 난수 개수는 10으로 하여라.)

모평균에 관한 유의성 검정

- 가설 $H_0 : \mu=0.5$, $H_1 : \mu \neq 0.5$


- 유의수준 $\alpha=0.05, n=10 $


- 검정통계량 $T=\frac{\bar{X}-\mu_0}{S/\sqrt{n}}\sim t(9)$ under $H_0$

```python
import scipy
from scipy.stats import bernoulli
import numpy as np
import scipy.stats as st
from scipy.stats import t

np.random.seed(0)
x=bernoulli.rvs(0.5,size=10)
print(x)

n=np.count_nonzero(x)
print(n)

t1=st.ttest_1samp(x,0.5)
print(round(t1.statistic,3))
print(round((np.mean(x)-0.5)/((np.std(x)*np.sqrt(10/9))/np.sqrt(10)),3))

print(round(t1.pvalue,3))
print(round(2*(1-t.cdf(t1.statistic,9)),3))
# 유의확률이 0.05보다 크므로 귀무가설을 기각할 수 없다.
```

## 2. 
실제 모수 $\mu=0.35$ 인 베르누이 확률 변수의 시뮬레이션을 통해 귀무 가설 $H_0:\mu=0.5$인 검정을 해보아라. (scipy.stats의 내장함수를 이용하되, 식으로 계산한 값과 비교하여라. 난수 개수는 100으로 하여라.)

- 가설 $H_0 : \mu=0.5$, $H_1 : \mu \neq 0.5$


- 유의수준 $\alpha=0.05, n=100$


- 검정통계량 $T=\frac{\bar{X}-\mu_0}{S/\sqrt{n}}\sim t(99)$ under $H_0$

```python
np.random.seed(0)
x=bernoulli.rvs(0.35,size=100)

n=np.count_nonzero(x)
print(n)

t1=st.ttest_1samp(x,0.5)
print(round(t1.statistic,3))
print(round((np.mean(x)-0.5)/((np.std(x)*np.sqrt(100/99))/np.sqrt(100)),3))

print(round(t1.pvalue,4))
print(round(2*(t.cdf(t1.statistic,99)),4))
# 유의확률이 0.05보다 작으므로 귀무가설을 기각할 수 있다.
```

## 4. 
a반과 b반이 만점이 200점인 시험을 동일하게 보았다고 하자. 두 반의 성적은 다음과 같았다.

a반 : 175, 168, 168, 190, 156, 181, 182, 175, 174, 179

b반 : 185, 169, 173, 173, 188, 186, 175, 174, 179, 180

두 반의 성적의 평균에 차이가 없는지에 대해 알아보려고 한다. 이에 필요한 검정을 실시하여라. (필요한 함수를 사용하되, 식으로 계산한 값과 비교하여라.)

(1) 등분산 검정

- 가설 $H_0 : \sigma_1^2=\sigma_2^2$, $H_1 : \sigma_1^2\neq \sigma_2^2$


- 검정통계량 $F=\frac{S_1^2}{S_2^2}\sim F(9,9)$ under $H_0$

```python
from scipy.stats import f

a = np.array([175, 168, 168, 190, 156, 181, 182, 175, 174, 179])
b = np.array([185, 169, 173, 173, 188, 186, 175, 174, 179, 180])

var_ratio = np.var(a)/np.var(b)
print(round(var_ratio,3))

print(round(1/f.ppf(0.975,9,9),3),round(f.ppf(0.975,9,9),3))
# 기각역의 경계

print(round(2*(1-f.cdf(var_ratio,9,9)),3))
# 양측 가설의 유의확률
# 귀무가설을 기각할 수 없다. 따라서 등분간 가정이 만족한다.
```

(2) 등분산 독립 이표본 검정

(1)의 결과에 따라 등분산 가정 하에 모평균 비교 검정을 할 수 있다.

- 가설 $H_0 : \mu_1=\mu_2$, $H_1 : \mu_1 \neq \mu_2$


- 합동표본분산 ${S_p}^2=\frac{(n_1-1)s1^2+(n_2-1)*s2^2}{n_1+n_2-2}$


- 검정통계량 $T=\frac{(\bar{x_1}-\bar{x_2})-0}{S_p \sqrt{\frac{1}{n_1}+\frac{1}{n_2}}}\sim t(n_1+n_2-2)$ under $H_0$


```python
S_p=np.sqrt((10*np.var(a)+10*np.var(b))/(10+10-2)) 
# Pooled standard deviation

tt=st.ttest_ind(a,b,equal_var=True) 
print(round(tt.statistic,3))

t1=(np.mean(a)-np.mean(b))/S_p/np.sqrt(1/10+1/10)
print(round(t1,3))
#검정통계량

print(round(tt.pvalue,3))
print(round(2*t.cdf(t1,10+10-2),3))
# 유의확률
# 귀무가설을 기각할 수 없다.
```

## 5. 
중간고사 이후 과외를 받은 10명의 학생의 중간고사 – 기말고사 점수 데이터를 가상으로 만들어서 비교해 보았다. 성적은 다음과 같다.

mid : 6, 20, 21, 22, 23, 22, 27, 25, 27, 28

final : 19, 22, 24, 24, 25, 25, 26, 26, 28, 32

과외 전후 성적에 변화가 있는지에 대한 검정을 실시하여라. (필요한 함수를 사용하되, 식으로 계산한 값과 비교하여라.)

대응표본 t test

- 가설 $H_0 : \mu_1=\mu_2$, $H_1 : \mu_1 \neq \mu_2$


- 유의수준 $\alpha=0.05, n=10$


- 검정통계량 $T=\frac{\bar{D}-0}{S_D/\sqrt{n}}\sim t(9)$ under $H_0$

```python
mid = np.array([16, 20, 21, 22, 23, 22, 27, 25, 27, 28])
final = np.array([19, 22, 24, 24, 25, 25, 26, 26, 28, 32])

tt=st.ttest_rel(mid,final)
print(round(tt.statistic,3))
# ttest_rel을 사용한 검정통계량 

d=np.mean(mid-final)
s=np.std(mid-final)*np.sqrt(10/9)

print(round(d/(s/np.sqrt(10)),3))
# 식을 이용하여 계산한 검정통계량. 둘의 결과는 같다.

print(round(tt.pvalue,3))
print(round(2*t.cdf(d/(s/np.sqrt(10)),9),3))
         
# 유의확률. 귀무가설을 기각할 수 있다. 과외를 받은 전과 이후의 평균 성적 차이는 통계적으로 유의미하다고 할 수 있다. 
```

- $df^*=\frac{({S_A}^2/n_A+{S_B}^2/n_B)^2}{\frac{1}{n_A-1}({S_A}^2/n_A)^2+\frac{1}{n_B-1}({S_B}^2/n_B)^2}$
